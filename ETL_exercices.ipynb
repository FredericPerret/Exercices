### Ex E5 ⭐⭐ — Fonction extract()
Encapsulez tout dans une fonction `extract(q1_path, q2_path, clients_path, produits_path) -> dict`
qui retourne un dictionnaire avec les 4 DataFrames harmonisés et consolidés.
La fonction doit afficher des logs à chaque étape.

```python
import logging

logging.basicConfig(level=logging.INFO, format="%(asctime)s | %(message)s", datefmt="%H:%M:%S")
logger = logging.getLogger("ETL")

def extract(q1_path, q2_path, clients_path, produits_path):
    res = {}
    try:
        for path in [q1_path, q2_path, clients_path, produits_path]:
            res[path] = pd.read_csv(path)
            logger.info("chargement de "+path+" réussi")
    except:
        logger.error("chargement de "+path+" a échoué")
        raise
    try:
        res[q2_path].rename(columns={"moyen_paiement":"mode_paiement"},inplace=True)
        logger.info("renommage de la colonne moyen_paiement de Q2 réussi")
    except:
        logger.error("renommage de la colonne moyen_paiement de Q2 a échoué")
        raise
    try:
        res[q1_path]["devise"] = "EUR"
        logger.info("ajout de la colonne devise dans Q1 réussi")
    except:
        logger.error("ajout de la colonne devise dans Q1 a échoué")
        raise
    try:
        res[q1_path]["source"] = "Q1_2024"
        res[q2_path]["source"] = "Q2_2024"
        logger.info("ajout de la colonne source dans Q1 et Q2 réussi")
    except:
        logger.error("ajout de la colonne source dans Q1 et Q2 a échoué")
        raise
    try:
        commandes_q1q2 = pd.concat([res[q1_path], res[q2_path]],ignore_index=True)
        logger.info("concaténation de Q1 et Q2 réussie")
    except:
        logger.error("concaténation de Q1 et Q2 a échoué")
        raise
    return (commandes_q1q2,res[clients_path],res[produits_path])
data = extract("data/commandes_Q1.csv","data/commandes_Q2.csv","data/clients.csv","data/produits.csv")
```

### Ex T8 ⭐⭐⭐ — Validation qualité
Écris une fonction `valider_pipeline(df) -> bool` qui :
1. Vérifie qu'il n'y a plus de NaN sur les colonnes critiques
2. Vérifie qu'il n'y a plus de doublons
3. Vérifie que tous les prix sont > 0
4. Vérifie que toutes les quantités sont > 0
5. Vérifie que montant_total == quantite × prix_unitaire (à 1 centime près)
6. Retourne True si tout est OK, False sinon avec les erreurs détaillées

```python
def valider_pipeline(df):
    err = []
    # Vérifie qu'il n'y a plus de NaN sur les colonnes critiques :
    for col in ("commande_id","client_id","produit_id","quantite","prix_unitaire","montant_total","date_commande","statut","mode_paiement"):
        nbnull = df[col].isnull().sum()
        if nbnull > 0:
            err.append("Colonne "+col+" : "+str(nbnull)+" NaN")
    # Vérifie qu'il n'y a plus de doublons :
    listDbl = df.value_counts("commande_id")>1
    for key in listDbl.keys():
        if listDbl[key]:
            err.append("Clé "+key+" en doublon")
    # Vérifie que tous les prix sont > 0 :
    for col in ("prix_unitaire","montant_total"):
        nbPrixNeg = (df[col]<0).sum()
        if nbPrixNeg > 0:
            err.append("Colonne "+col+" : "+str(nbPrixNeg)+" prix négatifs")
    # Vérifie que toutes les quantités sont > 0 :
    nbQuantNeg = (df["quantite"]<=0).sum()
    if nbQuantNeg > 0:
        err.append("Colonne quantite : "+str(nbQuantNeg)+" quantités négatives ou nulles")
    # Vérifie que montant_total == quantite × prix_unitaire (à 1 centime près) :
    for cmd in df.itertuples(index=False):
        if abs(cmd.montant_total - cmd.prix_unitaire * cmd.quantite) > 0.01:
            err.append("Commande "+cmd.commande_id+" : montant_total différent de prix_unitaire * quantite")
    # Retourne True si tout est OK, False sinon avec les erreurs détaillées :
    if len(err) > 0:
        print("Liste des erreurs :")
        for e in err:
            print(e)
        return False
    else:
        return True
valider_pipeline(data[0])
```
